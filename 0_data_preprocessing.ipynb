```python
# Basic setup
import pandas as pd
import numpy as np
import os

# Show all columns in outputs
pd.set_option('display.max_columns', None)

# Set dataset path
data_path = '../data/CICIDS2017/'
filename = 'Friday-WorkingHours-Afternoon-DDos.pcap_ISCX.csv'
file_path = os.path.join(data_path, filename)

print("Reading file from:", file_path)
```

```python
# Ensure pandas is imported
import pandas as pd

# Read your dataset
df = pd.read_csv(file_path)

# Strip all column names to remove extra spaces
df.columns = df.columns.str.strip()

# Print column names (optional but helpful for debugging)
print("Columns in dataset:")
print(df.columns)

# Safely print unique values from 'Label' column
if 'Label' in df.columns:
    print("\nUnique labels:")
    print(df['Label'].value_counts())
else:
    print("\nError: 'Label' column not found in the dataset. Please check the column names above.")
```

```python
# Drop columns that are all NaN or empty strings
df = df.dropna(axis=1, how='all')  # Drop fully-empty columns
df = df.loc[:, (df != ' ').any(axis=0)]  # Drop columns with only blank spaces

print("Shape after dropping empty columns:", df.shape)
```

```python
# Preview cleaned data
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder

# Strip column names again just in case
df.columns = df.columns.str.strip()

# Initialize encoder
le = LabelEncoder()
df['Label'] = le.fit_transform(df['Label'])

# Store class map for reference
label_mapping = dict(zip(le.classes_, le.transform(le.classes_)))
print("Label Encoding Mapping:\n", label_mapping)
```

```python
# Drop non-feature columns (e.g., Flow ID, Timestamp if present)
X = df.drop(['Label'], axis=1)
y = df['Label']

print("Features shape:", X.shape)
print("Target shape:", y.shape)
```

```python
# Fix invalid values in X
X = X.replace([np.inf, -np.inf], np.nan)
X = X.dropna()

# Keep y aligned with filtered X
y = y[X.index]

# Apply SMOTE
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)

print("\nAfter SMOTE:")
print("X:", X_resampled.shape)
print("y distribution:\n", pd.Series(y_resampled).value_counts())
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_resampled)

# Convert back to DataFrame
X_scaled = pd.DataFrame(X_scaled, columns=X.columns)

X_scaled.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y_resampled, test_size=0.2, random_state=42, stratify=y_resampled
)

print("Train shape:", X_train.shape)
print("Test shape:", X_test.shape)
```

```python
# Reshape to 3D for LSTM [samples, timesteps, features]
X_train_lstm = X_train.values.reshape(X_train.shape[0], 1, X_train.shape[1])
X_test_lstm = X_test.values.reshape(X_test.shape[0], 1, X_test.shape[1])

print("Reshaped X_train shape:", X_train_lstm.shape)
print("Reshaped X_test shape:", X_test_lstm.shape)
```

```python
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, LSTM, Conv1D, MaxPooling1D, Flatten, Dense, Dropout, concatenate

# Input layer
input_layer = Input(shape=(X_train_lstm.shape[1], X_train_lstm.shape[2]))

# LSTM branch
lstm_out = LSTM(64, return_sequences=True)(input_layer)
lstm_out = Flatten()(lstm_out)

# CNN branch
cnn_out = Conv1D(filters=64, kernel_size=1, activation='relu')(input_layer)
cnn_out = MaxPooling1D(pool_size=1)(cnn_out)
cnn_out = Flatten()(cnn_out)

# Merge both branches
merged = concatenate([lstm_out, cnn_out])
dense = Dense(64, activation='relu')(merged)
drop = Dropout(0.3)(dense)
output = Dense(1, activation='sigmoid')(drop)

# Build model
model = Model(inputs=input_layer, outputs=output)
model.summary()
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

history = model.fit(
    X_train_lstm, y_train,
    epochs=10,
    batch_size=256,
    validation_data=(X_test_lstm, y_test),
    verbose=1
)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Predict
y_pred = (model.predict(X_test_lstm) > 0.5).astype(int)

# Report
print(classification_report(y_test, y_pred))

# Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

```python
# Cell 16: Plot Training & Validation Accuracy/Loss

import matplotlib.pyplot as plt

# Accuracy
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Model Accuracy Over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid()
plt.show()

# Loss
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Model Loss Over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid()
plt.show()
```

```python
import tensorflow as tf
import numpy as np

# Wrap model in a TF function to compute gradients
loss_object = tf.keras.losses.BinaryCrossentropy()

def create_adversarial_pattern(input_data, input_label):
    with tf.GradientTape() as tape:
        tape.watch(input_data)
        prediction = model(input_data)
        loss = loss_object(input_label, prediction)

    # Get gradients of the loss w.r.t. input
    gradient = tape.gradient(loss, input_data)
    signed_grad = tf.sign(gradient)
    return signed_grad
```

```python
# Pick a subset of test samples
num_samples = 1000
X_adv = tf.convert_to_tensor(X_test_lstm[:num_samples])
y_adv = tf.convert_to_tensor(np.array(y_test[:num_samples]).reshape(-1, 1), dtype=tf.float32)

# Create perturbations
epsilon = 0.1  # Small tweak for evasion
perturbations = create_adversarial_pattern(X_adv, y_adv)
X_fgsm = X_adv + epsilon * perturbations
X_fgsm = tf.clip_by_value(X_fgsm, -3, 3)  # Keep values bounded (after normalization)
```

```python
# Predict on adversarial data
y_pred_fgsm = (model.predict(X_fgsm) > 0.5).astype(int)

# Evaluate
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

print("Adversarial FGSM Evaluation Report:")
print(classification_report(y_test[:num_samples], y_pred_fgsm))

# Confusion Matrix
cm = confusion_matrix(y_test[:num_samples], y_pred_fgsm)
sns.heatmap(cm, annot=True, fmt='d', cmap='Reds')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix on FGSM Adversarial Examples')
plt.show()
```
